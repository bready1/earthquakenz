```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
tt=np.linspace(0,10,10000)
sig1=np.sin(2*pi*0.1*tt+2)*1+np.sin(2*pi*10*tt+5.6)+np.sin(2*pi*tt+1.2)
# tt=np.linspace(-1,1,200)
sig1=np.sin(2*pi*tt*(1+np.random.randn(tt.size)*0.1*np.sin(tt)))
# sig1  = np.cos(2 * np.pi * 7 * tt) + signal.gausspulse(tt - 0.4, fc=2)

# widths=np.arange(1,31)
widths=np.logspace(-4,2,num=301)
```

```python
np.log10(0.1)
```

```python
plt.plot(tt,sig1)
```

```python
cmtmatr=signal.cwt(sig1,signal.ricker,widths)
plt.figure(figsize=(16,12))
plt.imshow(abs(cmtmatr)**2, extent=[tt[0], tt[-1], widths[0]/2/pi,widths[-1]/2/pi], aspect='auto',)
plt.yscale("log")
plt.colorbar()
```
